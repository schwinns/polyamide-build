```python
import numpy as np
import gromacs as gro
import matplotlib.pyplot as plt
```

```python
scaled = gro.fileformats.XVG('scaled_pdf.xvg')
unscaled = gro.fileformats.XVG('unscaled_pdf.xvg')
```

```python
x = scaled.array[0,:]
y = scaled.array[1,:]
plt.plot(x,y,label='scaled')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.legend();
```

```python
xvg = gro.fileformats.XVG('box_equil.xvg')
xvg
```

```python
print(xvg.error) # nm
print(xvg.tc/1000) # ns
```

```python
data = xvg.array
x = data[0,:]
y = data[1:,:]
x = x/1000 + 15
plt.plot(x,y[0,:],c='indianred',label='x-y box length')
#plt.plot(x,y[1,:],c='goldenrod',label='y box length')
plt.plot(x,y[2,:],c='forestgreen',label='z box length')
plt.xlabel('Time (ns)')
plt.ylabel('Box length (nm)')
plt.legend()
plt.title('Box Length at Equilibrium')
plt.savefig('box_equil.png')
```
